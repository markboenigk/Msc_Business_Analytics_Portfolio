## Problem 1

To solve that problem and find a solution, I classified the products into categories like pants and shirts using regular expressions. I then analyzed the products.


```python
# Categorize prodcuts into clothing categories like pants, shirts and underwear 
def categorize_clothes(df_sports):
    clothes_map = {
        'shirt': ['tshirt', 'shirt', 'polo', 'jersey', 'tank', 'top', 'tee'],
        'underwear': ['bra', 'boxer'],
        'socks': ['socks'],
        'pants': ['leggings', 'pant', 'pants', 'sweatpants', 'tights'],
        'shorts': ['shorts', 'short'],
        'jacket': ['jacket', 'tracksuit'],
        'sweater': ['sweater', 'sweatshirt', 'hoodie'],
        'vest': ['vest'],
        'fleece': ['fleece'],
        'shoes': ['shoe', 'shoes', 'cleat', 'slide']
    }
    
    df_sports['clothes_type'] = 'other'
    df_sports['name'] = df_sports['name'].str.lower()
    df_sports['name'] = df_sports['name'].str.replace("-", "")
    
    # search for keywords with regex
    for clothes_type, keywords in clothes_map.items():
        regex = r'\b({})\b'.format('|'.join(keywords))
        df_sports.loc[df_sports['name'].str.contains(regex), 'clothes_type'] = clothes_type

    # Group data 
    df_sports_grouped  = df_sports.groupby(by=["clothes_type"]).agg({"old_price": ["count","sum", "mean"], "current_price":["sum", "mean"], "saving" : ["sum", "mean"] }).reset_index()
    df_sports_grouped = df_sports_grouped.sort_values([('old_price', 'count')], ascending=False)
    df_sports_grouped["discount_in_perc"] =  round(((df_sports_grouped[('current_price', 'mean')] /  df_sports_grouped[('old_price', 'mean')]) - 1) * 100,2)
    df_sports_grouped["share_volume"] =  round((df_sports_grouped[('current_price', 'sum')] / df_sports_grouped[('current_price', 'sum')].sum()) * 100, 2)
    df_sports_grouped.columns = ["clothes_type", "product_count", "original_price_sum","original_price_avg", "current_price_sum", "current_price_avg","savings_sum", "savings_mean", "discounts_avg_percentage", "share_volume"]
    df_sports_grouped[['original_price_avg','current_price_avg']] = df_sports_grouped[['original_price_avg','current_price_avg']].round(2)
    df_sports_grouped = df_sports_grouped.reset_index(drop = True)
    
    
    return df_sports, df_sports_grouped

df_sports, df_sports_grouped = categorize_clothes(df_sports)

# Create Plot
clothes_plot = sns.barplot(data=df_sports_grouped, x='clothes_type', y='product_count', palette='mako')
clothes_plot.set_xticklabels(clothes_plot.get_xticklabels(), rotation=45)
clothes_plot.set_title('Number of items per cloth type')
clothes_plot.set_ylabel('Number of items')
clothes_plot.set_xlabel('Cloth type')

# add percentage labels above each bar
for p in clothes_plot.patches:
    height = p.get_height()
    clothes_plot.text(p.get_x() + p.get_width() / 2, height,\
         f'{p.get_height()/df_sports_grouped["product_count"].sum()*100:.1f}%', ha='center')

plt.show()
df_sports_grouped[['clothes_type', 'product_count', 'original_price_avg','current_price_avg','discounts_avg_percentage' ]]
```

After that, I identified countries that have a lower average temperature than the current countries, Germany, Austria, Switzerland, and France. I used boolean subsetting for that and merged tables to identify possible countries. 

```python
current_countries = ["Germany", "Switzerland", "Austria", "France"]

def purchasing_power_worldbank(current_countries):
    p_power = pd.read_csv("World Bank/countries-dataset-2020/Cost of living index by country 2020.csv")
    p_power = p_power[["Country", "Local Purchasing Power Index"]]
    p_power["Country"] = p_power["Country"].str.strip()

    avg_p_power_current_countries = p_power["Local Purchasing Power Index"][p_power["Country"].isin(current_countries)].mean()
    potential_countries = p_power[(p_power["Local Purchasing Power Index"] > avg_p_power_current_countries) \
                                  & (~p_power["Country"].isin(current_countries))].sort_values(["Local Purchasing Power Index"], ascending=False).reset_index(drop=True)
    return p_power, potential_countries #potential_countries

def get_weather_data(current_countries):
    df_temp = pd.read_csv("World Bank/environmental-variables-for-world-countries/World_countries_env_vars.csv")
    df_temp = df_temp[['Country', 'temp_mean_annual','temp_annual_range', 'temp_min_coldestMonth', 'temp_max_warmestMonth']].sort_values(["temp_mean_annual"])

    avg_temp_current_countries = df_temp["temp_mean_annual"][df_temp["Country"].isin(current_countries)].mean()
    potential_countries = df_temp[(df_temp["temp_mean_annual"] <= avg_temp_current_countries) \
                                  & (~df_temp["Country"].isin(current_countries))]
    return potential_countries

# Call dataframes from functions
weather_df = get_weather_data(current_countries)
countries_w_purchasing_power, potential_p_power_countries = purchasing_power_worldbank(current_countries)

# See which countries are colder and have a higher purchasing power than current countries
df_match_temp_purchasing_power = pd.merge(weather_df,potential_p_power_countries, on="Country", how="inner" )
df_match_temp_purchasing_power 
```

```python

def categorize_clothes(df_sports):
    clothes_map = {
        'shirt': ['tshirt', 'shirt', 'polo', 'jersey', 'tank', 'top', 'tee'],
        'underwear': ['bra', 'boxer'],
        'socks': ['socks'],
        'pants': ['leggings', 'pant', 'pants', 'sweatpants', 'tights'],
        'shorts': ['shorts', 'short'],
        'jacket': ['jacket', 'tracksuit'],
        'sweater': ['sweater', 'sweatshirt', 'hoodie'],
        'vest': ['vest'],
        'fleece': ['fleece'],
        'shoes': ['shoe', 'shoes', 'cleat', 'slide']
    }
    
    df_sports['clothes_type'] = 'other'
    df_sports['name'] = df_sports['name'].str.lower()
    df_sports['name'] = df_sports['name'].str.replace("-", "")
    
    # search for keywords with regex
    for clothes_type, keywords in clothes_map.items():
        regex = r'\b({})\b'.format('|'.join(keywords))
        df_sports.loc[df_sports['name'].str.contains(regex), 'clothes_type'] = clothes_type

    # Group data 
    df_sports_grouped  = df_sports.groupby(by=["clothes_type"]).agg({"old_price": ["count","sum", "mean"], "current_price":["sum", "mean"], "saving" : ["sum", "mean"] }).reset_index()
    df_sports_grouped = df_sports_grouped.sort_values([('old_price', 'count')], ascending=False)
    df_sports_grouped["discount_in_perc"] =  round(((df_sports_grouped[('current_price', 'mean')] /  df_sports_grouped[('old_price', 'mean')]) - 1) * 100,2)
    df_sports_grouped["share_volume"] =  round((df_sports_grouped[('current_price', 'sum')] / df_sports_grouped[('current_price', 'sum')].sum()) * 100, 2)
    df_sports_grouped.columns = ["clothes_type", "product_count", "original_price_sum","original_price_avg", "current_price_sum", "current_price_avg","savings_sum", "savings_mean", "discounts_avg_percentage", "share_volume"]
    df_sports_grouped[['original_price_avg','current_price_avg']] = df_sports_grouped[['original_price_avg','current_price_avg']].round(2)
    df_sports_grouped = df_sports_grouped.reset_index(drop = True)
    
    
    return df_sports, df_sports_grouped

df_sports, df_sports_grouped = categorize_clothes(df_sports)

df_sports_grouped
```

```python
# Analyse USA dataset 
# Drop columns, only select important ones 
def transform_sales(df_sales):
    df_sales = df_sales.drop(["Late_delivery_risk", "Customer Email", "Customer Fname","Customer Password", \
         "Customer Street", "Customer Zipcode"], axis=1)

    df_sales = df_sales[["Product Name","Category Name","Department Name","Sales", "Order Item Discount", "Order Item Discount Rate", "Order Item Profit Ratio"]]
    return df_sales


# Categorize products based on words into clothing categories 
def categorize_clothes_us(df_sales_usa):
    clothes_map = {
        'shirt': ['tshirt', 'shirt', 'polo', 'jersey', 'tank', 'top', 'tee'],
        'underwear': ['bra', 'boxer'],
        'socks': ['socks', 'sock', 'soc'],
        'pants': ['leggings', 'pant', 'pants', 'sweatpants', 'tights'],
        'shorts': ['shorts', 'short'],
        'jacket': ['jacket', 'tracksuit'],
        'sweater': ['sweater', 'sweatshirt', 'hoodie'],
        'vest': ['vest'],
        'fleece': ['fleece'],
        'shoes': ['shoe', 'shoes', 'cleat', 'slide'],
        'electronics': ['watch', 'monitor', 'laptop', 'camera', 'electronics', 'fitbit', 'activity tracker'],
        'equipment': ['putter', 'backpack', 'ball', 'balls', 'rip deck', 'gun', 'wedge', 'kayak', 'elliptical', 'bike']
    }
    df_sales1 = df_sales_usa
    df_sales1['clothes_type'] = 'other'
    df_sales1['Product Name'] = df_sales1['Product Name'].str.lower()
    df_sales1['Product Name'] = df_sales1['Product Name'].str.replace("-", "")
    
    for clothes_type, keywords in clothes_map.items():
        regex = r'\b({})\b'.format('|'.join(keywords))
        df_sales1.loc[df_sales1['Product Name'].str.contains(regex), 'clothes_type'] = clothes_type

    return df_sales1

# Create grouped dataframe with sales, sales price, discounts and profit per clothing categroy 
def group_us_sales(df_sales_usa_w_cloth_category):
    df = df_sales_usa_w_cloth_category
    df_grouped = df.groupby(['clothes_type']).agg({"Sales":["sum","mean"],"Order Item Discount Rate":"mean", "Order Item Profit Ratio": "mean"}).round(4).reset_index()
    df_grouped.columns = ["Clothes Type", "Total Sales", "Average Sales Price", "Average Discount Rate in %", "Average Profit in %"]
    df_grouped["Total Sales"] = (df_grouped["Total Sales"] / 1000).round(2)
    df_grouped = df_grouped.rename(columns={"Total Sales": 'Total Sales in Thousand'})
    df_grouped["Average Sales Price"] = df_grouped["Average Sales Price"].round(2)
    df_grouped[["Average Discount Rate in %", "Average Profit in %"]] =  df_grouped[["Average Discount Rate in %", "Average Profit in %"]] * 100
    return df_grouped

df_sales_usa = transform_sales(df_sales)
df_sales_usa_w_cloth_category = categorize_clothes_us(df_sales_usa)
grouped_us_sales = group_us_sales(df_sales_usa_w_cloth_category)
grouped_us_sales 

```
